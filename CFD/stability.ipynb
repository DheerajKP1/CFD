```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r_values = [0.8,1,1.2]
k_dx = np.linspace(0,2*np.pi,100)
plt.figure(figsize=(5,5))
colors = ['blue','red','green']
for i,r in enumerate(r_values):
    up_wind_ex = np.abs(1 - r + r*np.exp(1j*k_dx))
    up_wind_ap = 1 - r*0.5*(k_dx)**2
    up_ratio = up_wind_ap / up_wind_ex

    plt.plot(k_dx,up_ratio,color = colors[i], label=f"r={r}")
    plt.xlabel("k_dx")
    plt.ylabel("G_app/G_exact")
    plt.grid(True)
    plt.legend()
```

```python
r_values = [0.8,1,1.2]
k_dx = np.linspace(0,2*np.pi,100)
plt.figure(figsize=(5,5))
colors = ['blue','red','green']
for i, r in enumerate(r_values):
    lw_ex = np.abs( 1 - r**2 + 0.5*(r**2+r)*np.exp(1j * k_dx) + 0.5*(r**2-r)*np.exp(-1j * k_dx))
    lw_ap = np.sqrt((1 - 0.5*(r**k_dx)**2)**2+(r**k_dx)**2)

    lw_ratio = lw_ap/lw_ex

    plt.plot(k_dx,lw_ratio, color = colors[i], label=f"r={r}")
    plt.xlabel("k_dx")
    plt.ylabel("G_app/G_exact")
    plt.grid(True)
    plt.legend()
```
